```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import poisson
%matplotlib inline
```

```python
distribution = norm(loc=0, scale=1)
xs = np.linspace(-3,3, 100)
ys = distribution.pdf(xs)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.plot(xs, ys, '-')
ax.set_title('PDF norm(mu=0, sigma=1)', size=20)
ax.set_xlabel('x in standard deviations', size=15)
ax.set_ylabel('pdf(x)', size=15)
```

```python
distribution = norm(loc=0, scale=1)
xs = np.linspace(-3.0, 3.0, 100)
ys = distribution.cdf(xs)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.plot(xs, ys)
ax.set_title('CDF norm(mu=0, sigma=1)', size=20)
ax.set_xlabel('x in standard deviations', size=15)
ax.set_ylabel('cdf(x)', size=15)
```

```python
distribution = norm(loc=0, scale=1)
xs = np.linspace(-3.0, 3.0, 100)
cdf_ys = distribution.cdf(xs)
pdf_ys = distribution.pdf(xs)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.plot(xs, pdf_ys, color='b', linestyle=':', linewidth=5, label='PDF')
ax.plot(xs, cdf_ys, color='r', linestyle='-.', linewidth=5, label='CDF')
ax.set_title('CDF and PDF norm(mu=0, sigma=1)', size=20)
ax.set_xlabel('x in standard deviations', size=15)
ax.legend(loc='best')
```

```python
xs = [0, 1]
p = 0.1
pmf_ys = [bernoulli.pmf(k, p) for k in xs]
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(3, 3))
ax.set_title('Bernoulli PMF, p=0.1')
ax.set_xlabel('outocme (k)')
ax.set_ylabel('probability of outcome')
ax.set_ylim(0.0, 1.0)
ax.set_xlim(-1, 2)
ax.scatter(xs, pmf_ys, color='b', alpha=0.5, marker='*', s=250)
ax.vlines(xs, 0, pmf_ys, color='r', lw=5, alpha=0.5)
```

```python
n = 100
p = 0.2
xs = np.arange(binom.ppf(q=0.01, n=n, p=p), binom.ppf(q=0.99, n=n, p=p))
pmf_ys = binom.pmf(xs, n=n, p=p)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
ax.scatter(xs, pmf_ys, color='b', alpha=1.0, marker='^', s=250)
ax.vlines(xs, 0, pmf_ys, color='g', lw=5, alpha=0.5)
ax.set_title('Binomial PMF, n=100, p=0.1', size=20)
ax.set_xlabel('Number of successful trials (k)', size=15)
ax.set_ylabel('PMF(k)', size=15)
```

```python
n = 100
p = 0.1
xs = np.arange(binom.ppf(q=0.01, n=n, p=p), binom.ppf(q=0.99, n=n, p=p))
cmf_ys = binom.cdf(xs, n=n, p=p)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
ax.scatter(xs, cmf_ys, color='b', alpha=1.0, marker='^', s=250)
ax.vlines(xs, 0, cmf_ys, color='g', lw=5, alpha=0.5)
ax.set_title('Binomial CMF, n={}, p={}'.format(n, p), size=20)
ax.set_xlabel('Number of successful trials (success <= k)', size=15)
ax.set_ylabel('CMF(k)', size=15)
```

```python
mus = [10,20,30]
colors = ['g','c','b']


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
for mu, c in zip(mus, colors):
    xs = np.arange(poisson.ppf(q=0.001, mu=mu), poisson.ppf(q=0.999, mu=mu), 1)
    pmf_ys = poisson.pmf(xs, mu=mu)
    ax.scatter(xs, pmf_ys, color=c, alpha=1.0)
    ax.vlines(xs, 0, pmf_ys, color=c, lw=5, alpha=0.5)
    ax.set_title('Posisson PMF, mu={}'.format(mu), size=20)
    ax.set_ylabel('pmf(k)', size=15)
    ax.set_xlabel('k', size=15)
    #ax.set_ylim(0, pmf_ys.max() * 1.1)
```

```python
mu = 10
xs = np.arange(poisson.ppf(q=0.001, mu=mu), poisson.ppf(q=0.999, mu=mu), 1)
cmf_ys = poisson.cdf(xs, mu=mu)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
ax.scatter(xs, cmf_ys, color='r', alpha=1.0)
ax.vlines(xs, 0, cmf_ys, color='r', lw=5, alpha=0.5)
ax.set_title('Posisson CMF, mu={}'.format(mu), size=20)
ax.set_ylabel('cmf(k)', size=15)
ax.set_xlabel('k', size=15)
ax.set_ylim(0, cmf_ys.max() * 1.1)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(10, 10))
for mu, ax in zip(range(1, 13, 2), axs.flatten()):
    xs = np.arange(poisson.ppf(q=0.001, mu=mu), poisson.ppf(q=0.999, mu=mu), 1)
    pmf_ys = poisson.pmf(xs, mu=mu)
    ax.scatter(xs, pmf_ys, color='r', alpha=1.0, label='pmf(k)')
    ax.vlines(xs, 0, pmf_ys, color='r', lw=5, alpha=0.5)
    ax.set_title('Posisson PMF, mu={}'.format(mu))
    ax.set_ylim(0, pmf_ys.max() * 1.1)
    ax.legend(loc='best')
```
